```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/cleveland.csv", header=None)
print(df.shape)
df.head()
```

```python
df.columns = [
    "age",
    "sex",
    "cp",
    "trestbps",
    "chol",
    "fbs",
    "restecg",
    "thalach",
    "exang",
    "oldpeak",
    "slope",
    "ca",
    "thal",
    "target",
]
df.head()
```

```python
df.isna().mean()
```

```python
df["target"] = df.target.map({0: 0, 1: 1, 2: 1, 3: 1, 4: 1})
df["thal"] = df.thal.fillna(df.thal.mean())
df["ca"] = df.ca.fillna(df.ca.mean())
```

```python
# distribution of target vs age
df.age[df.target == 1].plot(kind="kde")
df.age[df.target == 0].plot(kind="kde")
plt.legend(["Disease", "No Disease"])
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Variation of Age for each target class")
plt.show()
```

```python
# barplot of sex vs age with hue = target
sns.barplot(x="sex", y="age", hue="target", data=df)
plt.show()
```
